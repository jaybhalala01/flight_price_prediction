```python
def chisquare(c1,c2):
    # Create the contingency matrix
    contingency_matrix = pd.crosstab(df[c1], df[c2])

    # Print the contingency matrix
    #print(contingency_matrix)

    from scipy.stats import chi2_contingency

    # Perform chi-square test
    statistic, p_value, dof, expected = chi2_contingency(contingency_matrix)

    # Print the results
    '''
    print("Chi-square statistic:", statistic)
    print("P-value:", p_value)
    print("Degrees of freedom:", dof)
    print("Expected frequencies:", expected)
    '''
    
    filename = '/home/growlt199/Downloads/manufacturing_project/model_training/data_categorical_column_pvalue.csv'
    
    with open(filename, 'a', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow([c1,c2,statistic,p_value,dof])  # Write the data row
    
    print(f'data loaded successfully for {c1} and {c2}..')    
```

```python
# Select columns with categorical data type
categorical_columns = df.select_dtypes(include=['object', 'category'])

# Get the column names
categorical_column_names = categorical_columns.columns

# Print the column names
#print(categorical_column_names)

# Iterate through the categorical column names
for i in range(len(categorical_column_names)):
    c1 = categorical_column_names[i]
    for j in range(i+1, len(categorical_column_names)):
        c2 = categorical_column_names[j]
        chisquare(c1, c2)
```
